```python
import numpy as np
import pandas as pd
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import normalize
```

```python
df= pd.read_csv("/content/Creditcard_data.csv")
df.head()
df=df.iloc[:,1:]
```

```python
Amount = normalize([df['Amount']])[0]
df['Amount'] = Amount
```

```python
df['Class'].value_counts()
```

```python
y=df.iloc[:,-1]
x=df.iloc[:,:-1]
x
```

```python
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(x,y)
y_resampled.value_counts()
```

```python
new_df=pd.concat([X_resampled,y_resampled],axis=1)
```

```python
new_df.head()
new_df['Class'].value_counts()
```

```python
## Simple random sampling
n = int((1.96*1.96 * 0.5*0.5)/(0.05**2))
sampled_df_1 = new_df.sample(n=n, random_state=42)
# sampled_df_1 = new_df.sample(frac=0.5, random_state=42)
sampled_df_1
```

```python
#systematic sampling
interval = 2

# perform systematic sampling
sampled_df_2 = new_df.iloc[::interval]
sampled_df_2
```

```python
##stratified sampling

strata = new_df.groupby('Class')
n = int((1.96*1.96 * 0.5*0.5)/((0.05)**2))
# sample 2 rows from each stratum
sampled_df_3 = strata.apply(lambda x: x.sample(n))

sampled_df_3
```

```python
#cluster sampling
# cluster sampling
import numpy as np

from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=10, random_state=42).fit(new_df)
cluster_assignments = kmeans.labels_

# Select the clusters you want to include in the sample
selected_clusters = [0, 2, 4, 5, 8]

cluster_series = pd.Series(cluster_assignments)

# Create the new DataFrame containing only the rows from the selected clusters
sampled_df_4 = new_df[cluster_series.isin(selected_clusters)]


# Print the resulting DataFrame
sampled_df_4
```

```python
#convenience sampling
sample_head=new_df.head(350)
sample_tail=new_df.tail(350)
sampled_df_5=pd.concat([sample_head,sample_tail])
sampled_df_5
```
